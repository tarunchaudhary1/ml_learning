```python
import matplotlib.pyplot as plt
numerical_features = ['milage', 'engine', 'price']
for feature in numerical_features:
  plt.figure(figsize=(8,4))
  sns.boxplot(x=df_train[feature])
  plt.title(f'Boxplot for {feature}')
  plt.show()
```

```python
# Calculate IQR for each feature
Q1 = df_train['milage'].quantile(0.25)
Q3 = df_train['milage'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f'Mileage outlier range: less than {lower_bound} or more than {upper_bound}')

df_train['milage'] = df_train['milage'].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))

Q1 = df_train['price'].quantile(0.25)
Q3 = df_train['price'].quantile(0.75)

IQR = Q3 - Q1

# Define lower and upper bound to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the data to remove outliers in the price column
filtered_df_train = df_train[(df_train['price'] >= lower_bound) & (df_train['price'] <= upper_bound)]


# Check the shape of data before and after outlier removal
print(f"Original data shape: {df_train.shape}")
print(f"Filtered data shape: {filtered_df_train.shape}")
df_train['price'] = df_train['price'].apply(lambda x:upper_bound if x> upper_bound else (lower_bound if x< lower_bound else x))
```

```python
sns.set(style="whitegrid")

numerical_features = ['milage', 'price']
for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.histplot(df_train[feature], kde=True, bins=30, color='blue')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df_train[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='milage', y='price', data=df_train, color='green')
plt.title('Scatter Plot of Mileage vs Price')
plt.xlabel('Milage')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='engine', y='price', data=df_train, color='red')
plt.title('Scatter Plot of Engine Size vs Price')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(10, 6))


plt.figure(figsize=(8, 4))
sns.countplot(x='fuel_type', data=df_train, palette='Set2')
plt.title('Fuel Type Count Distribution')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.show()

sns.pairplot(df_train[numerical_features], diag_kind='kde', corner=True)
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```
